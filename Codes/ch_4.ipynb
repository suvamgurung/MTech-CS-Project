```python
tc= pd.read_csv('/content/drive/MyDrive/Final (Completed)/training_corpus.csv')
train_data= list(tc.values)
```

```python
import ast
for i in range(0,len(train_data)):
    train_data[i][1]=ast.literal_eval(train_data[i][1])
type(train_data[0][1])
```

```python
labels= pd.read_csv('/content/drive/MyDrive/Final (Completed)/labels.csv')
#labels
```

```python
import spacy
#Training Existing NER Model
nlp_1= spacy.load('en_core_web_sm')
ner_1= nlp_1.get_pipe("ner") #Getting pipeline component

# Adding labels to the 'ner_1'
for label in labels:
    ner_1.add_label(label)

#Disable pipeline components you dont need to change
pipe_exceptions=["ner","trf_wordpiecer","trf_tok2vec"]
unaffected_pipes=[pipe for pipe in nlp_1.pipe_names if pipe not in pipe_exceptions]
```

```python
import random
from spacy.util import minibatch, compounding
from pathlib import Path
from spacy.training.example import Example
```

```python
#Training the model

import random
from spacy.util import minibatch, compounding
from pathlib import Path
from spacy.training.example import Example

with nlp_1.disable_pipes(*unaffected_pipes):
    for iteration in range(10):
        random.shuffle(train_data)
        losses={}
        #batch up the examples before every iteration
        batches= minibatch(train_data,size=compounding(4.0,32.0,1.001))
        for batch in batches:
            texts, annotations= zip(*batch)
            #print(len(texts),len(annotations))
            for i in range(0,len(texts)):
                #print(i,texts[i],annotations[i])
                doc = nlp_1.make_doc(texts[i])
                example = Example.from_dict(doc, annotations[i])
                nlp_1.update([example], losses=losses, drop=0.5)
                #print("losses",losses)
        print("Iteration",iteration)
```

```python
t= 'PMJAY scheme has PMJAY Beneficiary who goes to any EHCP\'s EHCP Registration Desk at an Empanelled Hospital'
d= nlp_1(t)
for ent in d.ents:
    print(ent)
```
