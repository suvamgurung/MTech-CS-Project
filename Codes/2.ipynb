`````python
from tenacity import (
    retry,
    stop_after_attempt,
    wait_random_exponential,
)  # for exponential backoff


@retry(wait=wait_random_exponential(min=1, max=60), stop=stop_after_attempt(6))
def completion_with_backoff(**kwargs):
    return openai.Completion.create(**kwargs)
`````

# `generate_sentence(entity_pairs):` To generate 5 sentences containing a pair of related entities.
`Input: DataFrame having two ent1 and ent2 columns`<br>
`Output: List`

`````python
def generate_sentences(keywords):
    sent_text=[]
    count=0
    for i in range(0,len(keywords)):
        if(count==1):
            print('Wait for 2 seconds to avoid -> [429: \'Too Many Requests\' or RateLimitError]...')#Waiting for 2 sec to be safe from 429: 'Too Many Requests' or RateLimitError Error of OpenAI
            time.sleep(2)#wait for 2 Seconds to further avoid RateLimit Error of Open AI
            count=0
        p='Create 10 sentence relating to Ayushman Bharat PMJAY with the following:'+keywords['ent1'][i]+','+keywords['ent2'][i]
        print('Request:',i,': [',keywords['ent1'][i],',',keywords['ent2'][i],']')
        response= completion_with_backoff(model="text-davinci-003", prompt=p, temperature=0.7, max_tokens=3970)
        sent_text.append(response['choices'][0]['text'])
        #print(sent_text)
        count+=1
    return (sent_text)
`````
